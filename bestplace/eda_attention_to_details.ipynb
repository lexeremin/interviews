### Замечание: 
Колонка с выборкой должна состоять только из 2х значений. Как можно заметить здесь имеются не определенные значения, сокращенные значения и значения, судя по всему, содержащие латиницу и кириллицу в одном слове. 

Можно использовать 2 способа чтобы привести к виду с двумя значениями:
1. Отфильтровать обущающие по `id` из выборки, остальные задать как тестовые.
2. Сначала исправить значения `-` на `nan`, оставить слово обучающая только с кириллицой, тест заменить на тестовая.
Первый способ более практичный, но с ним можно потерять информацию о том были ли в датасети заведомо ложные тестовые значения(заданы как тестовые, но их `id` присутствуют в обучающей выборке), поэтому далее будет рассмотрен второй способ.

Ниже приведем к виду с 3 значениями (обучающая, тестовая, nan)

```python
import re
re.search(r'[a-zA-Z]', unique_samples[-1]) 
```

В полученном массиве с уникальными значениями в выборке последнее слово начинается с латинской буквы `y` заменим на слово, состоящее полностью из кириллицы

```python
df1.loc[df1['Выборка'] == unique_samples[-1], 'Выборка'] = unique_samples[0]
df1.loc[df1['Выборка'] == unique_samples[2], 'Выборка'] = unique_samples[3]
df1.loc[df1['Выборка'] == unique_samples[4], 'Выборка'] = unique_samples[1]
```

```python
unique_samples = df1['Выборка'].unique()
unique_samples
```

### Замечание: 
Обучающая выборка содержит, значения помеченные в `df1` как `Тестовая`, это говорит о не точности предоставленных данных клиентом. Поскольку эти данные представлены в обучающей выборке они должны быть заданы как `Обучающие`. Аналогично сделаем и для `nan`.

```python
df1.loc[df1['id точки'].isin(intersected_id), 'Выборка']
```

```python
df1.loc[~df1['id точки'].isin(intersected_id), 'Выборка']
```

```python
df1.loc[df1['id точки'].isin(intersected_id), 'Выборка'] = "Обучающая"
```

```python
df1['Выборка'].unique()
```

Оставшиеся значения `nan` и `Обучающая`(`id` который отсутствуют в обучающей выборке) заменим на `Тестовая`, тк они осутствуют в `df2`

```python
df1.loc[~df1['id точки'].isin(intersected_id), 'Выборка'] = "Тестовая"
```

### Замечание:
В исходном датасете `df1`, исходя из выше результатов, имеются дублирующиеся `id точка`(42683), соответвенно, использование таких данных для расчета средней выручки не возможно, поскольку данные о выручке по обучающей выборке используют `id точка`, как внешний ключ для сопоставления данных. Из основного датафрейма `df1` не исключаем, тк по всем остальным параметрам это разные магазины, один из которых имеет бесплатную парковку в будет в включен в ответ на 3 вопрос

```python
valid_mall_ids = [i for i in valid_mall_ids if i != 42683]
```

```python
mini_malls_profit = df2.loc[df1['id точки'].isin(valid_mall_ids)]
mini_malls_profit = mini_malls_profit[mini_malls_profit.columns[1:]]
mini_malls_profit
```

Теперь неободимо отфильтровать лишние колонки, в получившемся датафрейме, чтобы упростить рассчет среднего значения выручки

```python
mini_malls_profit.columns = pd.to_datetime(mini_malls_profit.columns)
mini_malls_profit = mini_malls_profit.loc[:, mini_malls_profit.columns.year == 2016]
```

```python
mini_malls_profit = mini_malls_profit.dropna(how='all')
mini_malls_profit
```

```python
mini_malls_profit.info()
```

В полученном датафрейме видно, что у некоторых мини ТЦ отсутствует выручка в начале 2016, проверим есть ли среди них, открывшиеся в 2016 году.

```python
df1.loc[(df1['id точки'].isin(valid_mall_ids) & df1['Дата открытия'].dt.year == 2016)]
```

И если, те которые закрылись в 2016 году или ранее.

```python
df1.loc[(df1['id точки'].isin(valid_mall_ids) & df1['Дата закрытия'].dt.year)]
```
